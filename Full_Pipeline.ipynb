## 2. Document Classification && Data Extraction (ID number)

```python
def Personal_ID_Classifier(img_path, model):
    # read the image
    img = cv2.imread(img_path)
    # plt.imshow(img)
    # plt.show()

    # apply image preprocessing
    img = pre_process.orient(img)
    # img = pre_process.docTr_wrapper(img)

    # plt.imshow(img)
    # plt.show()

    # classify the image
    predicted_label = id_classification.classify_ID_img(img, model)

    # ---------------------------------------------------------------------------------------
    id_ocr = IdOCR()
    # check image for the next stage
    if len(predicted_label) == 1:
        if predicted_label[0] == 'ID':
            # move to the next stage
            try:
                name, pred_id = id_ocr.apply_ocr(img)
                message = f"Name: {name}\nID Number: {pred_id}"

            except CustomError as e:
                    message = e.message            

            else:
                # save the output to a file
                with open(f"output/id/{pred_id}.txt", "w") as file:
                    file.write("Name: "+ name + "\n")
                    file.write("ID Number: "+ pred_id + "\n")

        else:
            message = "Oops!! Wrong Image\nPlease upload a clear image for your National ID"
    else:
        try:
            id_index = predicted_label.index("ID")
            if id_index == 0:
                id_img = img[:img.shape[0]//2]
            else:
                id_img = img[img.shape[0]//2:img.shape[0]] 
        except ValueError:
            message = "Oops!! Wrong Image\nPlease upload a clear image for your National ID"
            return message
        
        # then move to the next stage
        try:
            name, pred_id = id_ocr.apply_ocr(id_img)
            message = f"Name: {name}\nID Number: {pred_id}"

        except CustomError as e:
                message = e.message  
                          
        else:
            # save the output to a file
            with open(f"output/id/{pred_id}.txt", "w") as file:
                file.write("Name: "+ name + "\n")
                file.write("ID Number: "+ pred_id + "\n")

    return message
    
```

## 3. Bachelor Certificate Classification

```python
# Use OCR.space
def recognize_text(image_path, api_key):
    payload = {
        'apikey': api_key,
        'language': 'ara',  # Set the language to Arabic
    }
    with open(image_path, 'rb') as image_file:
        files = {'file': image_file}
        response = requests.post('https://api.ocr.space/parse/image', data=payload, files=files)

    if response.status_code == 200:
        result = response.json()
        return result
    else:
        return None
    
def OCR(image_path):
    result = recognize_text(image_path, 'K83256134988957')
    if result and result['IsErroredOnProcessing'] is False:
        parsed_text = result['ParsedResults'][0]['ParsedText']
        # save the output to a file
        with open("output/Cert_information.txt", "w") as file:
            file.write("Information : \n"+ parsed_text + "\n")
            
        return parsed_text

#extract persons arabic names 
def extract_arabic_names(parsed_text):
    tagger = SequenceTagger.load("megantosh/flair-arabic-multi-ner")

    # Rest of your code

    # Perform NER on the parsed text
    sentences = flair.data.Sentence(parsed_text)
    tagger.predict(sentences)
    
    # Extract Arabic names
    arabic_names = []
    for entity in sentences.get_spans('ner'):
        if entity.tag == 'PER':
            arabic_names.append(entity.text)
    
    return arabic_names

#extract informations
def extract_information(ocr_result):

    ## Remove non-alphanumeric characters
    ocr_result = re.sub(r'[^a-zA-Z0-9\u0600-\u06FF\s:/]+', '', ocr_result)


    #print(ocr_result)
    # Define patterns for university name, person's name, and grade
    university_pattern = r'(جامعة [^ ]+|كلية [^ ]+|المعهد[^ ]+)'
    person_name_pattern = r'(?:السيد|الطالبة|الطالب):?\s*(?:/)?\s*([\u0600-\u06FF\s]+)'
    grade_pattern = r'(?:بتقدير عام|التقدير العام :|:بتقدير|تقدير عام:|تقدير:) ((?:جيد جدا|ممتاز مع مرتبة الشرف|امتياز مع مرتبة الشرف|جيد جدا|امتياز|امتياز مع مرتبة الشرف|جيد|جيد مع مرتبة الشرف))'
    #grade_pattern = r'(?<=التقدير العام).*?((?:جيد جدا|ممتاز مع مرتبة الشرف|امتياز مع مرتبة الشرف|جيد جدا|امتياز|جيد|جيد مع مرتبة الشرف))'



    # Initialize variables
    university_name = ''
    person_name = ''
    grade = ''

    # Extract information
    university_match = re.search(university_pattern, ocr_result)
    if university_match:
        university_name = university_match.group(0)

    person_match = re.search(person_name_pattern, ocr_result)
    if person_match:
        person_name = person_match.group(1)
    else:
      arabic_names = extract_arabic_names(ocr_result)
      person_name = arabic_names[0] if arabic_names else None

    grade_match = re.search(grade_pattern, ocr_result)
    if grade_match:
        grade = grade_match.group(1)



    # save the output to a file
    with open(f"output/grad_Cert_information.txt", "w") as file:
        file.write("university: "+ university_name + "\n")
        file.write("Name: "+ person_name + "\n")
        file.write("Grade: "+ grade + "\n")

    """# Print the extracted information
    if university_name:
        print(f"اسم الجامعة: {university_name}")
    if person_name:
        print(f"اسم الشخص: {person_name}")
    if grade:
        print(f"التقدير: {grade}")
    """
```

```python
def Bachelor_Cert_Classifier(img_path, model):
    img = cv2.imread(img_path)
    img = pre_process.orient(img)
    img = pre_process.docTr_wrapper(img)

    img_label = BC_Classifier.BC_Classifier(img, model)
    
    if img_label != 'Other':
        # Move to the next stage
        text = OCR(img_path)
        # print(text)
        
        extract_information(text)
    else:
        print("Oops!! Wrong Image\nPlease upload a clear image for your Bachelor Certificate")

    #return message
```

## 4. Military Certificate Classification

```python
import pytesseract
from PIL import Image
def Military_Cert_Classifier(img_path, model):
    img = cv2.imread(img_path)
    img = pre_process.orient(img)
    img = pre_process.docTr_wrapper(img)
    img_label = MC_Classification.MC_Classifier(img, model)

    if img_label != 'Other':
        # Move to the next stage
        message = prepro(img_path)
        
        if message == "not decided" :
            message = prepro(img_path,contrast=True)
    else:
        message = "Oops!! Wrong Image\nPlease upload a clear image for your Military Certificate"

    return message   


def catchNormal(text)  :
    if "نهائياً" in text or "النهائى"in text: 
        return "permenant"
    if "مؤقتا" in text or "المؤقت" in text : 
        return "temp"
    return "not decided"
def catch(text)  :
    normal  = catchNormal(text)
    if  normal != "not decided" :
        return normal
    text = text.split(" ")
    for word in text : 
        p = jaccard_similarity(word , "النهائي")
        t = jaccard_similarity(word , "مؤقت")
        if p > 0.5 :
            return "permenant"
        elif t > 0.5 :
            return "temp"

    return "not decided"



def jaccard_similarity(word1, word2):
    set1 = set(word1)
    set2 = set(word2)
    intersection = len(set1.intersection(set2))
    union = len(set1) + len(set2) - intersection
    similarity = intersection / union
    return similarity


def prepro(path , contrast = False , show_text=False) : 
    #image = Image.open(r"D:\ITI\GP\tesseract\scanner_20220802_114818.jpg") 
    image = cv2.imread(path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    if contrast == True :
        clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
        image = clahe.apply(image)

    tesseract_config = '--psm 6 --oem 3 -l ara'
    text = pytesseract.image_to_string(image, config=tesseract_config)
    if show_text == True :
        print(text)

    return catch(text)
```
